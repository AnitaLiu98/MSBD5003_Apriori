```
import itertools
import time
start = time.clock()
t0 = time.time()

def splitlines(x):
    y = ' '.join(x.split())
    return y
    
def verifyCandidates(transation, candidates) :
    for c in candidates :
        if (c.issubset(transation)):
             yield (c, 1)

def prune(transactions,Ck,support):
    # CSups candidate and supports
    CSups = {}
    for transaction in transactions:
        for candi in map(frozenset,Ck):
            if candi.issubset(transaction):
              if not candi in CSups:
                  CSups[candi] =1
              else:
                  CSups[candi] +=1

                  
    numRecords = len(transactions)
    
    # 满足最小支持度的频繁项集和频率
    # FreSups frequent itemset and support
    FreSups = {}
    CurFre = []
    for candi in CSups:
        support_candi = CSups[candi]/numRecords   
        if support_candi >= support:
            CurFre.append(candi)
        FreSups[candi] = support_candi
     
    return CurFre, FreSups
  
def genCk(Lk, k):
    Ck = [] 
    for i in range(len(Lk)):
        for j in range(i+1, len(Lk)):
            L1 = list(set(Lk[i]))[: k-2]
            L2 = list(set(Lk[j]))[: k-2]
            L1.sort()
            L2.sort()
            if L1 == L2:
                Ck.append(frozenset(Lk[i] | Lk[j]))
    return Ck  
  
iteration = []
with open('/dbfs/FileStore/tables/retail.txt') as f:
  for line in f:
    iteration.append(line.strip(' \n').split(' '))
print(iteration[0])
#iteration =['1 3 4 5','6 8 9 10','11 23 1 3']
frqtemp=[]
support = 0.005

numRecords = len(iteration)
freqThre = numRecords * support
#print ("numRecords：" + str(numRecords))
freqThre = numRecords * support
#print ("frequency threshold：" + str(freqThre))
transactions = iteration
#transactions = list(map(lambda x : x.split(' '),splitedline))
                 
         
#Create candidate_1
candidate_1 = list(set(list(itertools.chain.from_iterable(transactions))))#distinct
candidate_1 = list(map(lambda c: [c] ,candidate_1))
#print(candidate_1[0:10])
CurFre, FreSups = prune(transactions,candidate_1,support)
frqtemp.append(CurFre)
preFreSets = CurFre
  
         
#Create CK LK
K =200 #迭代次数
for r in range(2,K):
  #print ('round'+str(r)+':\n')
  #print (preFreSets)
  if len(preFreSets)==0:
    break
  candidates = genCk(preFreSets, r)
  CurFre, FreSups= prune(transactions,candidates,support)
  frqtemp.append(CurFre)
  preFreSets = CurFre
print (frqtemp[:-1])

end = time.clock()
print (end - start, "seconds process time")
print (time.time() - t0, "seconds wall time")
```
